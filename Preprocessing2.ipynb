```python
# Using Pandas
avg_age=df.Age.mean()
avg_StudyHours=df.StudyHours.mean()
avg_Attendance=df.Attendance.mean()
avg_Marks=df.Marks.mean()
#freq_Grade=df.Grade.dropna().mode()[0]   # [0]:retrives the 1st mode from the resulting series
freq_City=df.City.dropna().mode()[0]

df.Age.replace(np.nan,avg_age, inplace = True)
df.StudyHours.replace(np.nan,avg_StudyHours, inplace = True)
df.Attendance.replace(np.nan,avg_Attendance, inplace = True)
df.Marks.replace(np.nan,avg_Marks, inplace = True)
#df.Grade.replace(np.nan,freq_Grade, inplace = True)
df.City.replace(np.nan,freq_City, inplace = True)
```

```python
df.loc[df['Grade'].isna() & (df['Marks'] >= 90), 'Grade'] = 'A'
df.loc[df['Grade'].isna() & (df['Marks'] >= 75) & (df['Marks'] < 90), 'Grade'] = 'B'
df.loc[df['Grade'].isna() & (df['Marks'] >= 60) & (df['Marks'] < 75), 'Grade'] = 'C'
df.loc[df['Grade'].isna() & (df['Marks'] < 60), 'Grade'] = 'D'
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['StudyHours','Attendance','Marks']] = scaler.fit_transform(df[['StudyHours','Attendance','Marks']])
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder  # string data to numerical
encoder=OneHotEncoder()
df.City=encoder.fit_transform(df[['City']]).toarray()
df.head()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df.Gender=encoder.fit_transform(df[['Gender']])
df.head()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df.Grade=encoder.fit_transform(df[['Grade']])
df.head()
```
