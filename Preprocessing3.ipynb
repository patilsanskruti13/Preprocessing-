```python
df.drop_duplicates(inplace=True)
```

```python
avg_age=df.Age.mean()
avg_Salary=df.Salary.mean()
avg_Experience=df.Experience.mean()
avg_PerformanceRating=df.PerformanceRating.mean()

df.Age.replace(np.nan,avg_age, inplace = True)
df.Salary.replace(np.nan,avg_Salary, inplace = True)
df.Experience.replace(np.nan,avg_Experience, inplace = True)
df.PerformanceRating.replace(np.nan,avg_PerformanceRating, inplace = True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Age','Salary']] = scaler.fit_transform(df[['Age','Salary']])
df.head()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df.PerformanceRating=encoder.fit_transform(df[['PerformanceRating']])
df.head()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df.Gender=encoder.fit_transform(df[['Gender']])
df.head()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df.Department=encoder.fit_transform(df[['Department']])
df.head()
```
